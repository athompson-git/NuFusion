# Neutrino rates from fusion chain reactions in Tokamak, FRC, ICF reactors

```python
import pynucastro as pyrl
rl = pyrl.ReacLibLibrary()

rho = 1e-10 # g/cm^3 - typical for a fusion reactor
T = 1.0e8 # Kelvin - ITER temp ~ 10x sun


```

## Rates for $D + T \to ^{4}\rm{He} + n$ reactors

```python
nuc_list = ["p", "n", "d", "t", "he3", "he4"]
lib = rl.linking_nuclei(nuc_list)
pynet = pyrl.PythonNetwork(libraries=[lib])


comp_DT = pyrl.Composition(pynet.get_nuclei())

for k in comp_DT:
    if k == pyrl.Nucleus("d"):
        comp_DT[k] = 0.5
    elif k.raw == "t":
        comp_DT[k] = 0.5
    else:
        comp_DT[k] = 0.0

comp_DT.normalize()

pynet.evaluate_rates(rho, T, comp_DT)
```

```python
fig = pynet.plot(rho=rho, T=T, comp=comp_DT, hide_xp=True, node_size=600, curved_edges=True)
```
